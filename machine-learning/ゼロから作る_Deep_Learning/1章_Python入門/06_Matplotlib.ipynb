06 Matplotlib
============

* `Matplotlib`：グラフ描画のためのライブラリ

    * グラフの描画や、データの可視化が簡単に行える

## 1. 単純なグラフの描画

* グラフを描画するためには、`matplotlib`の`pyplot`というモジュールを利用する

* 例)$\sin$関数を描画する例を見てみる

```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline

# データの作成
x = np.arange(0, 6, 0.1) # 0から6まで0.1刻みで生成
y = np.sin(x)

# グラフの描画
plt.plot(x, y)
```

* ここでは、NumPyの`arange`メソッドによって、[0, 0.1, 0.2, ..., 5.8, 5.9]というデータを生成した

    * これを$x$とした
    
* この$x$の各要素を対象に、NumPyの$\sin$関数(`np.sin()`)を適用し、`x`、`y`のデータ列を`plt.plot`メソッドに与え、グラフを描画した

## 2. pyplotの機能

* $\cos$関数に加えて、タイトルや$x$軸ラベル名の描画など、`pyplot`の他の機能も利用してみる

```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline

# データの作成
x = np.arange(0, 6, 0.1) # 0から6まで0.1刻みで生成
y1 = np.sin(x)
y2 = np.cos(x)

# グラフの描画
plt.plot(x, y1, label="sin")
plt.plot(x, y2, linestyle="--", label="cos")
plt.xlabel("x")
plt.ylabel("y")
plt.title("sin & cos")
plt.legend()
```

## 3. 画像の表示

* また、画像の読み込みは`matplotlib.image`モジュールの`imread()`を利用することができる

```python
import matplotlib.pyplot as plt
from matplotlib.image import imread

img = imread('/Users/kunii.sotaro/Desktop/spring8-and-sacla_FZCGMkQ.jpg')
plt.imshow(img)
```
