03 数値微分
=========

* `勾配法`：勾配の情報を使って、進む方向を決める

## 1. 微分

* 微分とは、ある瞬間の変化の量を表したものであり、以下の数式で定義される

\begin{eqnarray}
\frac{df(x)}{dx} = 
\lim_{h \to 0} \frac{f(x+h)-f(x)}{h} 
\end{eqnarray}

* 実際に、関数の微分を求めるプログラムを実装する

```python
# 悪い実装例
def numerical_diff(f, x):
    h = 10e-50
    return (f(x + h) - f(x)) / h
```

* ただし、Pythonでは`10e-50`とすると`丸め誤差`という問題に直面する

    * これは、小数の小さな範囲に置いて数値が省略されること


```python
import numpy as np

np.float32(1e-50)
```

* ここでは、微小な値`h`として$10^{-4}$を用いる

* 次の改善ポイントとして、関数`f`の差分を考える

    * `x + h`と`x`の間での関数`f`の差分を計算しているが、$(x+h)$と$x$の間の傾きに対応してしまう
    
* そのため、$(x+h)$と$(x-h)$での関数$f$の差分を計算することで、誤差を計算する

    * その前後の差分を計算することから、`中心差分`と呼ぶ

* 改善点を基に、数値微分の実装を行う

```python
def numerical_diff(f, x):
    h = 1e-4
    return (f(x + h) - f(x - h)) / (2*h)
```

## 2. 数値微分の例

* 上の数値微分を用いて、簡単な関数を微分してみる

\begin{eqnarray}
y = 0.01 x^2 + 0.1 x
\end{eqnarray}

* Pythonで実装すると、以下のようになる

```python
def function_1(x):
    return 0.01 * x **2 + 0.1 * x
```

* 続いて、この関数を描画する

```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline

x = np.arange(0.0, 20.0, 0.1)
y = function_1(x)
plt.xlabel('x')
plt.ylabel('f(x)')
plt.plot(x, y)
```

* ここで、この関数の微分を、`x=5`と`x=10`の時で、それぞれ計算する

```python
numerical_diff(function_1, 5)
```

```python
numerical_diff(function_1, 10)
```

* ここで、計算した微分の解は以下の通りとなる

\begin{eqnarray}
\frac{df(x)}{dx} = 0.02x+0.1
\end{eqnarray}

* `x=5`の時、`0.2`

* `x=10`の時、`0.3`

* 上の数値微分の結果を使って、その数値微分の値を傾きとする直線をグラフにプロットする

```python
def numerical_diff(f, x):
    h = 1e-4 # 0.0001
    return (f(x+h) - f(x-h)) / (2*h)


def function_1(x):
    return 0.01*x**2 + 0.1*x 


def tangent_line(f, x):
    d = numerical_diff(f, x)
    print(d)
    y = f(x) - d*x
    return lambda t: d*t + y
     
x = np.arange(0.0, 20.0, 0.1)
y = function_1(x)
plt.xlabel("x")
plt.ylabel("f(x)")

tf = tangent_line(function_1, 5)
y2 = tf(x)

plt.plot(x, y)
plt.plot(x, y2)
```

```python
tf = tangent_line(function_1, 10)
y2 = tf(x)

plt.plot(x, y)
plt.plot(x, y2)
```

## 3. 偏微分

* 次の関数を考える

\begin{eqnarray}
f(x_0, x_1) = x_0^2 + x_1^2
\end{eqnarray}

* この式は、Pythonで実装すると以下のようになる

```python
def function_2(x):
    return x[0]**2 + x[1]**2
```

* ここでは、引数にNumPy配列が入力されることを想定する

* この関数をグラフとして描画すると、以下のようになる

```python
from mpl_toolkits.mplot3d import Axes3D

def main():
    x0 = np.arange(-2, 2, 0.1)
    x1 = np.arange(-2, 2, 0.1)
    X0, X1 = np.meshgrid(x0, x1)
    Y = X0 ** 2 + X1 ** 2

    fig = plt.figure()
    ax = Axes3D(fig)
    ax.plot_wireframe(X0, X1, Y)
    
if __name__ == '__main__':
    main()
```

* ここで、上の式の微分を求める

* ここで、「どの変数に対しての微分か？」ということを区別する(`偏微分`)

```python
# 問1：x_0=3, x_1=4の時のx_0に対する偏微分
def function_tmp1(x0):
    return x0*x0 + 4.0**2.0
```

```python
numerical_diff(function_tmp1, 3.0)
```

```python
# 問1：x_0=3, x_1=4の時のx_1に対する偏微分
def function_tmp2(x1):
    return 3.0**2.0  + x1*x1
```

```python
numerical_diff(function_tmp2, 4.0)
```
