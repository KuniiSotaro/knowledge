04 勾配
======

* ここでは、$x_0$と$x_1$の偏微分をまとめて計算することを考える

* `勾配`：全ての変数の偏微分を、ベクトルとしてまとめたもの

\begin{eqnarray}
(\frac{\partial f}{\partial x_0}, \frac{\partial f}{\partial x_1})
\end{eqnarray}

* `勾配`は、以下のように実装することができる

```python
import numpy as np

def numerical_gradient(f, x):
    h = 1e-4  # 0.0001
    grad = np.zeros_like(x)
    
    for idx in range(x.size):
        tmp_val = x[idx]
        x[idx] = float(tmp_val) + h
        fxh1 = f(x)  # f(x+h)
        
        x[idx] = tmp_val - h 
        fxh2 = f(x)  # f(x-h)
        grad[idx] = (fxh1 - fxh2) / (2*h)
        
        x[idx] = tmp_val  # 値を元に戻す
        
    return grad
```

* ここで、`np.zeros_like(x)`は、`x`と同じ形状の配列で、その要素が全て`0`の配列を生成する

* `numerical_gradient(f, x)`関数は、引数の`f`は関数、`x`はNumPy配列であるとして、NumPy配列の`x`の各要素に対して数値微分を求める

* ここで、この関数を用いて実際に勾配を計算する

```python
def function_2(x):
    return x[0]**2 + x[1]**2
```

```python
numerical_gradient(function_2, np.array([3.0, 4.0]))
```

```python
numerical_gradient(function_2, np.array([0.0, 2.0]))
```

```python
numerical_gradient(function_2, np.array([3.0, 0.0]))
```

* これらの勾配を、以下の図で表してみる

```python
import matplotlib.pylab as plt
from mpl_toolkits.mplot3d import Axes3D
%matplotlib inline


def _numerical_gradient_no_batch(f, x):
    h = 1e-4  # 0.0001
    grad = np.zeros_like(x)
    
    for idx in range(x.size):
        tmp_val = x[idx]
        x[idx] = float(tmp_val) + h
        fxh1 = f(x)  # f(x+h)
        
        x[idx] = tmp_val - h 
        fxh2 = f(x)  # f(x-h)
        grad[idx] = (fxh1 - fxh2) / (2*h)
        
        x[idx] = tmp_val  # 値を元に戻す
        
    return grad


def numerical_gradient(f, X):
    if X.ndim == 1:
        return _numerical_gradient_no_batch(f, X)
    else:
        grad = np.zeros_like(X)
        
        for idx, x in enumerate(X):
            grad[idx] = _numerical_gradient_no_batch(f, x)
        
        return grad


def function_2(x):
    if x.ndim == 1:
        return np.sum(x**2)
    else:
        return np.sum(x**2, axis=1)


def tangent_line(f, x):
    d = numerical_gradient(f, x)
    print(d)
    y = f(x) - d*x
    return lambda t: d*t + y


if __name__ == '__main__':
    x0 = np.arange(-2, 2.5, 0.25)
    x1 = np.arange(-2, 2.5, 0.25)
    X, Y = np.meshgrid(x0, x1)
    
    X = X.flatten()
    Y = Y.flatten()

    grad = numerical_gradient(function_2, np.array([X, Y]).T).T

    plt.figure()
    plt.quiver(X, Y, -grad[0], -grad[1],  angles="xy",color="#666666")
    plt.xlim([-2, 2])
    plt.ylim([-2, 2])
    plt.xlabel('x0')
    plt.ylabel('x1')
    plt.grid()
    plt.draw()
```

* 上の図のように、向きを持ったベクトルとして図示される

* `勾配`は、関数$f(x_0, x_1)$の「一番低い場所(最小値)」を指している

    * また、「一番低い場所」から遠く離れれば離れるほど、矢印の大きさも大きくなる

* より正確に言うならば、各場所において**関数の値を最も減らす方向**となる

## 1. 勾配法

* 機械学習の問題の多くは、学習の際に最適なパラメータを探索する

    * ここで最適なパラメータとは、損失関数が最小値をとるときのパラメータの値
    
    * `勾配法`では、勾配をうまく利用して関数の最小値を探そうという目的

* ここでの注意点は、各地点において関数の値を最も減らす方向を示すのが`勾配`であること

    * そのため、勾配が指す先が本当に関数の最小値なのかどうかを保証することはできない
    
    * しかし、その方向に進むことで関数の値を最も減らすことができるので、最小値の場所を探す問題においては、勾配の情報を手がかりに進む方向を決めるべき

* ここで、`勾配法`が登場する

    * この方法では、現在の場所から勾配方向に一定の距離だけ進む
    
    * そして、移動した先でも同様に勾配を求め、またその勾配方向へ進むというように、繰り返し勾配方向へ移動する
    
    * このように勾配方向へ進むことを繰り返すことで、関数の値を徐々に減らす
    
* この方法は機械学習の最適化問題でよく使われる手法であり、NNでの学習では勾配法が用いられる

* `勾配法`は、以下の数式で表すことができる

    * $\eta$：更新の量
    
\begin{eqnarray}
x_0 = x_0 - \eta \frac{\partial f}{\partial x_0}
\end{eqnarray}

\begin{eqnarray}
x_1 = x_1 - \eta \frac{\partial f}{\partial x_1}
\end{eqnarray}

* ここで、NNの学習において$\eta$は、`学習率`と呼ぶ

    * 1回の学習で、どれだけ学習すべきか、どれだけパラメータを更新するか、ということを決める

* 上の式では、1回の更新式を示しており、このステップを繰り返し行う

    * つまり、ステップごとに上の式のように変数の値を更新していき、そのステップを何度か繰り返すことで徐々に関数の値を減らしていく
    
    * また、変数の値が増えても同様に行う

* ただし、学習率の値は前もって何らかの値に決める必要がある

    * NNの学習では、学習率の値を変更しながら、正しく学習しているかどうか、確認作業を行うのが一般的

* ここで、勾配降下法(勾配法)をPythonで実装する

```python
def gradient_descent(f, init_x, lr=0.01, step_num=100):
    x = init_x
    
    for i in range(step_num):
        grad = numerical_gradient(f, x)
        x -= lr * grad
        
    return x
```

* 引数の`f`は最適化したい関数、`init_x`は初期値、`lr`は学習率、`step_num`は勾配法による繰り返しの数とする

* 関数の勾配は、`numerical_gradient(f, x)`で求めて、その勾配に学習率を掛けた値で更新する処理を`step_num`で指定された回数繰り返す

```python
# f(x_0, x_1)= x_0^2+x_1^2の最小値を勾配法で求める
def function_2(x):
    return x[0]**2 + x[1]**2
```

```python
init_x = np.array([-3.0, 4.0])
gradient_descent(function_2, init_x=init_x, lr=0.1, step_num=100)
```

* ここでは、初期値を`(-3.0, 4.0)`として、勾配法を使って最小値の探索を開始する

    * 最終的な結果は`(-6.11110793e-10,  8.14814391e-10)`となり、ほとんど`(0, 0)`に近い結果となる

* ここで、この図を以下に描画した

```python
import matplotlib.pylab as plt
%matplotlib inline

def numerical_gradient(f, X):
    if X.ndim == 1:
        return _numerical_gradient_no_batch(f, X)
    else:
        grad = np.zeros_like(X)
        
        for idx, x in enumerate(X):
            grad[idx] = _numerical_gradient_no_batch(f, x)
        
        return grad


def gradient_descent(f, init_x, lr=0.01, step_num=100):
    x = init_x
    x_history = []

    for i in range(step_num):
        x_history.append( x.copy() )

        grad = numerical_gradient(f, x)
        x -= lr * grad

    return x, np.array(x_history)


def function_2(x):
    return x[0]**2 + x[1]**2

init_x = np.array([-3.0, 4.0])    

lr = 0.1
step_num = 20
x, x_history = gradient_descent(function_2, init_x, lr=lr, step_num=step_num)

plt.plot( [-5, 5], [0,0], '--b')
plt.plot( [0,0], [-5, 5], '--b')
plt.plot(x_history[:,0], x_history[:,1], 'o')

plt.xlim(-3.5, 3.5)
plt.ylim(-4.5, 4.5)
plt.xlabel("X0")
plt.ylabel("X1")
```

* しかし、学習率は大きすぎても小さすぎても良い結果にならない

```python
# 学習率が大きすぎる例：lr=10.0
init_x = np.array([-3.0, 4.0])
gradient_descent(function_2, init_x=init_x, lr=10.0, step_num=100)
```

```python
# 学習率が小さすぎる例：lr=1e-10
init_x = np.array([-3.0, 4.0])
gradient_descent(function_2, init_x=init_x, lr=1e-10, step_num=100)
```

* 上の結果のように、学習率が大きすぎると大きな値へと発散されるが、小さすぎるとほとんどが更新されない

> `ハイパーパラメータ`：学習率のようなパラメータのこと
>
> 人の手によって設定されるパラメータなので、色々な値で繰り返してうまく学習できるケースを探す必要がある

## 2. ニューラルネットワークに対する勾配

* ニューラルネットワークの学習においても、勾配を求める必要がある

* ここで言う勾配とは、重みパラメータに関する損失関数の勾配

    * 例)形状が$2 \times 3$の重み$W$だけを持つニューラルネットワークがあり、損失関数を$L$で表す時を考える
    
    * 実際に数式に表すと、以下の式で表される
    
\begin{eqnarray}
W =
\begin{pmatrix}
w_{11} & w_{12} & w_{13} \\
w_{21} & w_{22} & w_{23} 
\end{pmatrix}
\end{eqnarray}

\begin{eqnarray}
\frac{\partial L}{\partial W} = 
\begin{pmatrix}
\frac{\partial L}{\partial w_{11}} & \frac{\partial L}{\partial w_{12}} & \frac{\partial L}{\partial w_{13}} \\
\frac{\partial L}{\partial w_{21}} & \frac{\partial L}{\partial w_{22}} & \frac{\partial L}{\partial w_{23}}
\end{pmatrix}
\end{eqnarray}

* $\frac{\partial L}{\partial W}$の各要素は、それぞれの要素に関する偏微分から構成される

    * 例)1行1列目の要素である$\frac{\partial L}{\partial w_{11}}$は、$w_{11}$を少し変化させると損失関数$L$がどれだけ変化するかを表す
    
    * ここで、$\frac{\partial L}{\partial W}$の形状は$W$と同じである

* 実際に、簡単なニューラルネットワークを例にして、実際に勾配を求める実装を行う

    * そのため、`simpleNet`と言うクラスを実装する

```python
import sys, os
sys.path.append(os.pardir)  # 親ディレクトリのファイルをインポートするための設定
import numpy as np
from common.functions import softmax, cross_entropy_error
from common.gradient import numerical_gradient


class simpleNet:
    def __init__(self):
        self.W = np.random.randn(2,3)

    def predict(self, x):
        return np.dot(x, self.W)

    def loss(self, x, t):
        z = self.predict(x)
        y = softmax(z)
        loss = cross_entropy_error(y, t)

        return loss
```

* `simpleNet`クラスだが、これは形状が$2 \times 3$の重みパラメータを1つだけインスタンス変数として持つ

    * また、2つのメソッドがあり、一つは予測するためのメソッド`predict(x)`、もう一つは損失関数の値を求めるためのメソッド`loss(x, t)`
    
    * ここで、引数の`x`には入力データ、`t`には正解ラベルが入力されるものとする

```python
net = simpleNet()

# 重みパラメータ
print(net.W) 
```

```python
x = np.array([0.6, 0.9])
p = net.predict(x)
print(p)
```

```python
# 最大値のインデックス
np.argmax(p)
```

```python
# 正解ラベル
t = np.array([0, 0, 1])
net.loss(x, t)
```

* 続いて、勾配を求めてみる

* `numerical_gradient(f, x)`を使って勾配を求める

```python
def f(W):
    return net.loss(x, t)
```

```python
dW = numerical_gradient(f, net.W)
print(dW)
```

* `numerical_gradient(f, x)`の引数`f`は関数、`x`は関数`f`への引数

    * ここでは`net.W`を引数に取り、損失関数を計算する新しい関数`f`を定義する
    
    * そして、その新しく定義した関数を、`numerical_gradient(f, x)`に渡す

* `numerical_gradient(f, net.W)`の結果は`dW`となり、形状が$2 \times 3$の2次元配列となる

    * `dW`の中身を見ると、$\frac{\partial L}{\partial W}$の$\frac{\partial L}{\partial w_{11}}$はおおよそ0.06となる
    
    * これは、$w_{11}$を$h$だけ増やすと、損失関数の値は$0.66h$だけ増加することを表す

* なお、`lambda`を使うと以下の記法となる

```python
f = lambda w: net.loss(x, t)
dW = numerical_gradient(f, net.W)
```

* ニューラルネットワークの勾配を求めれば、あとは勾配法にしたがって、重みパラメータを更新するのみとなる
